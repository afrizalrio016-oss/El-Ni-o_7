```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

file_nc = "cmems_mod_glo_phy_my_0.083deg_P1M-m_1780477828622.nc"

ds = xr.open_dataset(file_nc)

periods = [
    (" Pra El Niño Super 2014",
     "2014-01-01",
     "2014-12-31"),

    ("El Niño Super 2015-2016",
     "2015-01-01",
     "2016-12-31"),

    ("Pasca El Niño Super 2017",
     "2017-01-01",
     "2017-12-31"),

    ("Pra El Niño 2022",
     "2022-01-01",
     "2022-12-31"),

    ("El Niño 2023-2024",
     "2023-01-01",
     "2024-12-31"),

    ("Pasca El Niño 2025",
     "2025-01-01",
     "2025-12-31"),

    ("Rata-rata Arus 10 Tahun",
     "2015-01-01",
     "2024-12-31")
]

for idx, (judul, start, end) in enumerate(periods, start=1):


    u = ds["uo"].sel(
        time=slice(start, end)
    ).isel(depth=0).mean("time")

    v = ds["vo"].sel(
        time=slice(start, end)
    ).isel(depth=0).mean("time")

    u = u.fillna(0)
    v = v.fillna(0)

    lon = ds.longitude.values
    lat = ds.latitude.values

    lon2d, lat2d = np.meshgrid(lon, lat)

    speed = np.sqrt(
        u.values**2 +
        v.values**2
    )

    mean_speed = np.mean(speed)
    max_speed = np.max(speed)
    min_speed = np.min(speed)

    fig, ax = plt.subplots(
        figsize=(12, 9),
        dpi=300
    )

    levels = np.linspace(
        0,
        np.max(speed),
        30
    )

    cf = ax.contourf(
        lon,
        lat,
        speed,
        levels=levels,
        cmap="YlGnBu",
        extend="max"
    )

    ax.contour(
        lon,
        lat,
        speed,
        levels=15,
        colors="white",
        linewidths=0.4,
        alpha=0.5
    )

    skip = 2

    q = ax.quiver(
        lon2d[::skip, ::skip],
        lat2d[::skip, ::skip],
        u.values[::skip, ::skip],
        v.values[::skip, ::skip],
        color="black",
        scale=8,
        width=0.003
    )

    ax.quiverkey(
        q,
        0.90,
        0.04,
        0.2,
        "0.2 m/s",
        labelpos="E"
    )

    cbar = plt.colorbar(
        cf,
        ax=ax,
        pad=0.02
    )

    cbar.set_label(
        "Kecepatan Arus (m/s)",
        fontsize=12
    )

    info = (
        f"Mean : {mean_speed:.3f} m/s\n"
        f"Max  : {max_speed:.3f} m/s\n"
        f"Min  : {min_speed:.3f} m/s"
    )

    ax.text(
        0.02,
        0.97,
        info,
        transform=ax.transAxes,
        fontsize=11,
        va="top",
        bbox=dict(
            boxstyle="round",
            facecolor="white",
            alpha=0.85
        )
    )

    ax.set_title(
        judul,
        fontsize=18,
        fontweight="bold",
        color="#1f7a4d",
        loc="left"
    )

    fig.suptitle(
        "Rata-rata Arus Permukaan (~0.49 m) | CMEMS GLOBAL_MULTIYEAR_PHY",
        fontsize=11,
        color="dimgray"
    )

    ax.set_xlabel(
        "Longitude (°E)",
        fontsize=14
    )

    ax.set_ylabel(
        "Latitude (°N)",
        fontsize=14
    )

    ax.grid(
        linestyle="--",
        alpha=0.4
    )

    plt.tight_layout()

    nama_file = (
        f"{idx:02d}_"
        + judul.replace(" ", "_")
        + ".png"
    )

    plt.savefig(
        nama_file,
        dpi=300,
        bbox_inches="tight"
    )

    plt.close()

    print("Tersimpan:", nama_file)

print("Selesai membuat 7 visualisasi.")
```
